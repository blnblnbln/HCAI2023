Usaremos python para crear y leer un archivo en formatio "binario".
Primero leemos el archivo creado anteriormente "datos.txt"

```python
import numpy as np
asdf = np.loadtxt("datos.txt", usecols=[1], unpack=True) 
```

```python
#La cantidad de datos se ve con:
asdf.shape
```

```python
#Crear un array con np
q = np.arange(asdf.shape[0])
```

Ahora queremos guardar los datos:

```python
import struct
```

```python
ej = 5
```

```python
ej_bin = struct.pack("i", ej)
```

```python
type(ej_bin)
```

```python
#Al imprimirlo muestra datos de memoria en lenguaje máquina
print(ej_bin)
#Veo 4 bytes, estás usando 4 bytes para 1 entero 
```

```python
ej2 = 2.2
ej2_bin = struct.pack("f", ej2)
print(ej2_bin)
```

Ahora podemos guardar nuestro array de la clase anterior de datos aleatorios

```python
with open("holabuenas.bln", "wb") as f:
    f.write(ej_bin)
    f.write(ej2_bin)
```

Como resultado vemos que el archivo tiene 8 bytes, 4 para el entero y 4 para el flotante. Ahora podemos leer el archivo:

```python
with open("holabuenas.bln", "rb") as f:
    number_bin = f.read()
```

```python
#if: int y float
numbers = struct.unpack("if", number_bin)
```

```python
numbers
```

# Archivo masivo en binario
Ahora guardaremos el array del archivo original:

```python
#asdf datos.txt, q:index np.arange 
with open("datos.bin", "wb") as f:
    for i in q:
        bin_data = struct.pack("f", i)
        f.write(bin_data)
    for i in asdf:
        bin_data = struct.pack("f", i)
        f.write(bin_data)
```
